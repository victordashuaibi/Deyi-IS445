#Read csv file to dataframe, mark missing values as NA
df = pd.read_csv("crash_data.csv", na_values = ['unknown','Unknown','UNKNOWN','0'])
df.head()

```python
df.shape
```

```python
#Remove unecessary variables
df = df.drop(['Report Number',
            'Local Case Number',
            'Off-Road Description',
            'Municipality',
            'Related Non-Motorist',
            'Non-Motorist Substance Abuse',
            'Person ID',
            'Circumstance',
            'Vehicle ID',
            'Driverless Vehicle',
            'Parked Vehicle',
            'Equipment Problems',
            'Vehicle First Impact Location',
            'Vehicle Second Impact Location',
            'Road Name',
            'Cross-Street Name'], axis='columns')
df.head()
```

```python
#Check number of null values
df.isna().sum()
```

```python
#Check percent of null values
df.isna().sum()*100/len(df)
```

```python
#Drop all null values to prevent them from affecting visualization results
df = df.dropna().reset_index(drop = True)
```

```python
#Check duplicates
df.duplicated().sum()
```

```python
#Drop duplicates
df = df.drop_duplicates().reset_index(drop = True)
```

**Relationship between location and vehicle**

```python
df['Vehicle Body Type'].value_counts()
```

```python
#Visualization of number of crashed of each vehicle body type
plt.figure(figsize=(20,10))
sns.barplot(x = df['Vehicle Body Type'].index, y = df['Vehicle Body Type'])
plt.title('Vehicle Body Types')
plt.xlabel('Count')
plt.ylabel('Vehicle Body Types')
plt.show()
```

```python
vehicle_body_type = df.groupby('Vehicle Body Type').count()
vehicle_body_type.reset_index(inplace=True)
vehicle_body_type['Number of Crash'] = vehicle_body_type['Agency Name']
vehicle_body_type = vehicle_body_type[['Vehicle Body Type','Number of Crash']]
vehicle_body_type.head()
```

```python
#Visualization of number of crashed of each vehicle body type
plt.figure(figsize=(25,10))
sns.lineplot(x=vehicle_body_type['Vehicle Body Type'], y=vehicle_body_type['Number of Crash'], data = vehicle_body_type)
plt.xlabel('Vehicle Body Types')
plt.ylabel('Count')
plt.title('Number of Crashes of Each Vehicle Body Type')
plt.xticks(rotation=45)
plt.show()
```
